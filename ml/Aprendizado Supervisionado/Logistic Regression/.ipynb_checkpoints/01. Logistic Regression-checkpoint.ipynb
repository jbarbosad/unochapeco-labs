# Predictive Models

Machine Learning fundamentally differentiates itself from classical statistics by not assuming that the data comes from a specific model. Hence Machine Learning can, justifiably, try different models on a given dataset to eventually pick the best one.

Obviusly, to do this we need to define what it means to be "*best*". Different **Model Performance Measures** exist and any of these can be used to compare models - largely depending on the context and the kind of output.

- Regression outputs are continuouos numbers.

- Classification outputs are either:


    - Class output (from algorithms like SVM and KNN); or
    - Probability outputs (from algorithms like Logistic Regression and Random Forest).
    

# Logistic Regression

[**Logistic Regression**](https://christophm.github.io/interpretable-ml-book/logistic.html) models the probabilities for classification problems with **two** possible outcomes. It is an **extension of the Linear Regression** model for **Classification** problems.

**Classification** is again a Supervised Learning algorithm, meaning that there is a Dependent Variable that we are trying to undestand or predict, but this dependent variable is no longer a continuous number. Now, in Classification, it is a "*either or*", or "*belongs to one of categories*". For example: "*pass or fail*", "*yes or no*", "*one or zero*", "*event happening or event not happening*". Predicting if a student will pass or fail an exam is a classification problem, while trying to predict the student's grade is a regression problem.

## Why do we need Logistic Regression?

**Logistic Regression** uses a set of independent variables to predict a dependent variable which is categorical/discrete whereas **Linear Regression** is used where the dependent variable is numeric and continuous. Linear Regression cannot be used to predict probabilities because we can not restrict its output between $0$ and $1$.

Logistic Regression is a supervised learning algorithm that estimates the **log of odds** for an event which can be used to predict the probability of the occurance of that event. So, instead of fitting a straight line or hyperplane, the logistic regression model uses the **Logistic Function** to squeeze the output of a linear equation to between $0$ and $1$, in a Sigmoid function.

And it looks like this:

![Logistic Function](imgs/sigmoid.png)

The predicted probabilities can be converted to classes based on the **Threshold** (the default Threshold is $0.5$).

Linear Regression fits:

$$y = a + bx$$

Logistic Regression fits some function $f()$ of top of the regression.

$$y = f(a + bx)$$

The function $f()$ is non-linear, it is a **Sigmoid Function**. And there are several types of Sigmoid Function, but the two most popular kinds of S curves are:

$$\text{Logistic Regression} \to y = \frac{1}{1 + e^{-(a+bx)}}$$

$$\text{Probit Regression} \to y = \Phi(x)$$

![Probit vs Logistic Regression](imgs/normal_logit.png)

For the **Probit Regression** the letter Phi ( $\Phi$ ) stands for the **Cummulative Distribution Function** for the Normal Distribution.

We will focus on the **Logistic Regression Function**, also known as **Logit Function**. The **Logit Function** estimates the **Log of Odds**, where by Odds we mean the probability of getting a positive result ( $y$ ) divided by the probability of failing to get a positive result ( $1 - y$ ), which is called the **Odds Ratio**.

$$\text{Odds Ratio} = \frac{y}{1 - y} \to (y : 1 - y)$$

**Odds Ratio** is the probability of True ($1$) divided by the probability of False ($0$). And log of that is linear.

Being that said, we can easily find  an equivalent equation for the Logit Function.

$$\log\left({\frac{y}{1-y}}\right) = a + bx \to y = \frac{1}{1 + e^{-(a+bx)}}$$

**Note:** The **Logit Function** can also be represented in a different form, as below.

$$y = \frac{1}{1 + e^{-(a+bx)}} = \frac{e^{a+bx}}{e^{a+bx} + 1}$$

## Best Fit Logit Curve

In Linear Regression we find the best line by minimizing the **Sum of Squared Residuals** ( $\sum e_i^2$ ). This unfortunately will not work in Logistic Regression.

We are no longer fitting the linear equation \[ $y = a + bx$ \], but fitting a sigmoid function \[ $y = f(a+bx)$ \]. Because the sigmoid function is non-linear, once you square it up then things start to become complicated. The minimization problem is no longer easy, becoming a non-convex problem. Meaning that if we try to solve this problem, we might get multiple answers, or multiple curves and we might not know which one is the best curve.

Instead, for the Logit Function, we choose to minimize the **Log Loss** or **Cross-Entropy**.

$$\text{Log Loss } = -y \log(\hat{y}) - (1 - y) \log(1 - \hat{y})$$

$$y \to \text{observed}$$
$$\hat{y} \to \text{predicted}$$

We want the observations ( $y$ ) and predictions ( $\hat{y}$ ) to be as close as possible. In Classifications, the observations ( $y$ ) are always either $0$ or $1$ and the predictions ( $\hat{y}$ ) are numbers between $0$ and $1$. So we can look at the formula as conditionals.

$$\text{if }y = 1 \to -\log(\hat{y})$$
$$\text{if }y = 0 \to -\log(1 - \hat{y})$$

This way, the correct predictions will benefit the Log Loss Function (bring it close to $0$) and the wrong prediction will hurt the Log Loss Function (raise it to large numbers).

## Logistic Regression: Pros and Cons

### Advantages

- A classification model that does give probabilities.
- Easily extended to multiple classes (Multinomial Regression).
- Quick to train and very fast at classifying unknown records.

### Disadvantages

- Constructs linear boundaries.
- Assumes that variables are independent (e.g. does not include interaction terms).
- Interpretation of coefficients is difficult.

## Confusion Matrix

A **Confusion Matrix** is a tabular representation of the Predicted *vs* Actual classes. We can use it to assess the performance of a Logistic Regression model.

- **True Positives (TP):** The model predicted the class as positive, and it is actually positive.
- **True Negative (TN):** The model predicted the class as negative, and it is actually negative.
- **False Positive (FP):** The model predicted the class as positive, but it is actually negative (also known as **Type I Error**).
- **False Negative (FN):** The model predicted the class as negative, but it is actually positive (also known as **Type II Error**).

![Confusion Matrix](imgs/confusion_matrix.png)

Metrics that are often computed from a Confusion Matrix for a binary classifier:

- **Accuracy:** is the proportion of correct predictions over total predictions.
- **Precision:** is the proportion of positive identification that is actually correct.
- **Recall (or Sensitivity, or True Positive Rate TPR):** is the proportion of Total Positives that were correctly identified.
- **Specificity (or True Negative Rate TNR):** is the proportion of the Total Negatives correctly identified.

> **Note on Precision:** A model that produces no False Positives has a Precision of $1.0$.<br>
> **Note on Recall:** A model that produces no False Negatives has a Recall of $1.0$.

The formulas to calculate those metrics are:

$$\text{Accuracy} = \frac{TP + TN}{ TP + FP + TN + FN }$$

$$\text{Precision} = \frac{TP}{TP + FP}$$

$$\text{Recall} = \frac{TP}{TP + FN}$$

$$\text{Specificity} = \frac{TN}{TN + FP}$$
<br><br>

There is also a negative measure, the opposite of Accuracy called **Classification Error Rate**, which is the percentage of errors, the sum of Type I (FP) and Type II (FN) Errors.

$$\text{Classification Error Rate} = \frac{FP + FN}{ TP + FP + TN + FN }$$
<br><br>

Since **Accuracy** is the opposite of **Error Rate**:

$$\text{Accuracy} = 1 - \text{Error Rate}$$

## Performance Measures

### Accuracy

**Accuracy** is simply the overall percentage (%) of correct predictions, and can be very high even for very useless models.

![Accuracy](imgs/accuracy.png)

In the above example the Accuracy will be 98%, even if we predict all patients do not have cancer.

### Precision *vs* Recall

**Precision** should be used when we want to minimize False Positives, such as where the loss of resources is high. **Recall** should be used when we want to minimize False Negatives, such as where the loss of opportunity is high.

To fully evaluate the effectiveness of a model, you must examine **both** Precision and Recall. Unfortunately, Precision and Recall are often in tension. That is, improving Precision tipically reduces Recall and vice versa.

Various metrics have been developed that rely on both Precision and Recall. For example, the **Precision-Recall curve** and the **F1 Score**.

### Precision-Recall Curve

**Precision-Recall** is a useful measure of success of prediction when the classes are imblanced. The **Precision-Recall curve** shows the tradeoff between Precision and Recall for different thresholds. It can be used to select optimal threshold as required to improve the model performance.

![Precision-Recall Curve](imgs/precision-recall_curve.png)

In this example, Precision and Recall are the same when the threshold is $0.4$. If we want higher Precision, we can increase the threshold. If we want higher Recall, we can decrease the threshold.

### F1 Score

**F1 Score** (or **F-score**, or even **F-measure**) is a measure that takes into account both Precision and Recall. F1 Score is the harmonic mean of Precision and Recall. Therefore, this score takes both False Positives and False Negatives into account.

$$\text{F1 Score} = \frac{2 * \text{Precision} * \text{Recall}}{\text{Precision} + \text{Recall}}$$

The highest possible value of an **F1 Score** is $1$, indicating perfect Precision and Recall, and the lowest possible value is $0$.

## ROC Curves and AUC

For Classification problems with probability outputs, a Threshold can convert probability outputs to classifications. This choice of Threshold can change the Confusion Matrix.

As the Threshold changes, a plot of the False Positive Rate (FPR) *vs* True Positive Rate (TPR) is called the **ROC Curve** (Receiver Operating Characteristic).

### ROC Curve

The **ROC Curve** (or **Receiver Operating Characteristic Curve**) is a graph showing the performance of a classification model at all classification thresholds. This curve plots two parameters:

- True Positive Rate
- False Positive Rate

**True Positive Rate (TPR)** is a synonym for Recall and therefore is defined as follows:

$$TPR = \frac{TP}{TP+FN}$$

**False Positive Rate (FPR)** is the opposite of the Specificity and is defined as follows:

$$FPR = \frac{FP}{FP+TN}$$

A ROC curve plots **TPR** *vs* **FPR** at different classification thresholds. Lowering the classification threshold classifies more items as positive, thus increasing both False Positives and True Positives. The following figure shows a typical ROC Curve.

![ROC Curve](imgs/roc_curve.png)

To compute points in a ROC Curve, we could evaluate a Logistic Regression model many times with different classifications thresholds, but this would be inefficient. Fortunately, there is an efficient sorting-based algorithm that can provide this information for us, called **AUC**.

### Area Under the Curve (AUC)

**AUC** stands for **Area Under the (ROC) Curve**. That is, AUC measures the entire two-dimensional area underneath the entire ROC Curve from $(0, 0)$ to $(1, 1)$.

![AUC: Area Under the Curve](imgs/auc.png)

AUC provides an aggregate measure of performance across all possible classification thresholds. One way of interpreting AUC is as the probability. that the model ranks a random positive example more highly than a random negative example. For example, given the following examples, which are arranged from left to right in ascending order of Logistic Regression predictions:

![Output of Logistics Regression model](imgs/log_reg_mod.png)

AUC represents the probability that a random positive (green) example is positioned to the right of a random negative (red) example.

AUC ranges from $0$ to $1$. A model whose predictions are 100% wrong has an AUC of $0.0$; one whose predictions are 100% correct has an AUC of $1.0$. An AUC of $0.5$ represents random results, a diagonal line.

AUC is desirable for the following two reasons:

- AUC is **scale-invariant**: it measures how well predictions are ranked, rather than their absolute values.

- AUC is **classification-threshold-invariant**: it measures the quality of the model's prediction irrespective of what classification threshold is chose.

However, both there reasons come with caveats, which may limit the usefulness of AUC in certain use cases:

- **Scale invariance is not always desirable.** For example, sometimes we really do need well calibrated probability outputs, and AUC will not tell us about that.

- **Classification-threshold invariance is not always desirable.** In cases where there are wide disparities in the cost of False Negatives (FN) *vs* False Positives (FP), it may be critical to minimize one type of classification error. For example, when doing cancer detection, we likely want to prioritize minimizing False Negatives (even if that results in a significant increase of False Positives). AUC is not a useful metric for this type of optimization.

## Pima Indians Diabetes Case study

Diabetes is one of the most frequent diseases worldwide and the number of diabetic patients is growing over the years. The main cause of diabetes remains unknown, yet scientists believe that both genetic factors and environmental lifestyle play a major role in diabetes. 

Individuals with diabetes face a risk of developing some secondary health issues such as heart diseases and nerve damage. Thus, early detection and treatment of diabetes can prevent complications and assist in reducing the risk of severe health problems. 
Even though it's incurable, it can be managed by treatment and medication.

Researchers at the Bio-Solutions lab want to get better understanding of this disease among women and are planning to use machine learning models that will help them to identify patients who are at risk of diabetes. 

We will use logistic regression to model the "Pima Indians Diabetes" data set. In particular, all patients here are females at least 21 years old of Pima Indian heritage. This model will predict which people are likely to develop diabetes.

## Data Description:

* **Pregnancies**: Number of times pregnant
* **Glucose:** Plasma glucose concentration over 2 hours in an oral glucose tolerance test
* **BloodPressure:** Diastolic blood pressure (mm Hg)
* **SkinThickness:** Triceps skinfold thickness (mm)
* **Insulin:** 2-Hour serum insulin (mu U/ml)
* **BMI:** Body mass index (weight in kg/(height in m)^2)
* **Pedigree:** Diabetes pedigree function - A function that scores likelihood of diabetes based on family history.
* **Age:** Age in years
* **Class:** Class variable (0: the person is not diabetic or 1: the person is diabetic)

### Import the Libraries

```python
# Read and manipulate data
import numpy as np
import pandas as pd

# DataViz
import matplotlib.pyplot as plt
import seaborn as sns

# Split data
from sklearn.model_selection import train_test_split

# Build linear model for statistical analysis and prediction
import statsmodels.stats.api as sms
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm
from statsmodels.tools.tools import add_constant

# Get different metric scores
from sklearn import metrics
from sklearn.metrics import accuracy_score, roc_curve, confusion_matrix, roc_auc_score
```

### Read the data

```python
data = pd.read_csv('pima-indians-diabetes.csv')
```

### Checking for null values in the data

```python
data.isnull().values.any()
```

- We don't have missing values in the data.

### Distribution of variables in the data

```python
sns.set_style('darkgrid')
data.hist(figsize=(15,10))
plt.show()
```

- We can see that `Glucose`, `BloodPressure`, `SkinThickness`, `Insulin`, `BMI` have observations with $0$ values, which is not possible. These are data entry errors and we should fix it.
- We will replace $0$ values in all the columns except `Pregnancies` and `Class` with the `NaN` (this will help us deal withh the missing values elegantly).
- We will impute missing values (`NaN`) with the mean.

```python
# Columns to replace 0 with NaN
cols = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'Pedigree']
```

```python
# Replace 0 with NaN
data[cols] = data[cols].replace(0, np.nan)
```

```python
# Check missing values again
data.isna().sum()
```

```python
# Filling in the missing values with the mean
data[cols] = data[cols].fillna(data[cols].mean())
```

```python
# Check missing values again
data.isna().sum()
```

- We have imputed all the data entry errors with the Mean values

```python
# Check the distributions again
data.hist(figsize=(15,10))
plt.show()
```

### Distribution of the Target Variable

```python
data['Class'].value_counts()
```

```python
data['Class'].value_counts(normalize=True)
```

- The number of observations where the person is not diabetic are 500, which is 65% of the total data available.

### Splitting Data into Train and Test

```python
X = data.drop(['Class'], axis=1)
y = data['Class']
```

```python
# Add constant to the X variables
X = add_constant(X)
```

```python
# Create dummies
X = pd.get_dummies(X, drop_first=True) # Not necessary (?)
```

```python
# Split data into Train and Test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                   test_size=0.30,
                                                   random_state=1,
                                                   stratify=y) # maintains the original distribution
```

**Note:** the `stratify=y` argument maintains the original distribution of the classes (y) in the target variable while splitting the data into Train and Test sets.

### Fitting Logistic Regression model

```python
# Training the model
logit = sm.Logit(y_train, X_train.astype(float))
lg = logit.fit()
```

```python
# Print the Logistic Regression Summary
print(lg.summary())
```

- The coefficients of the Logistic Regression model are in terms of Log of Odds $\log(\text{odd})$.

- The interpretation of p-values ( `P > |z|` ) is similar to that in Linear Regression.

- As we do not have the simple R-squared ( $R^2$ ) like before, we have to look at the **Confusion Matrix** to get a good sense of how our model is performing.

### Checking the Model Performance

```python
# Predict on Training set
# Default Threshold is 0.5, if predicted probability is greater
# than 0.5 the observation will be classified as 1

pred_train = lg.predict(X_train) > 0.5
pred_train = np.round(pred_train)
```

### Confusion Matrix for the Training set

```python
cm = confusion_matrix(y_train, pred_train)

# Plot the Confusion Matrix (CM)
plt.figure(figsize=(7, 5))
sns.heatmap(cm, annot=True, fmt='g')
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.show()
```

#### The Confusion Matrix

- **True Positives (TP):** A person has diabetes and the model predicted person has diabetes. $\to$ $114$ TPs.

- **True Negatives (TN):** A person does not have diabetes and the model predicted person does not have diabetes. $\to$ $308$ TNs.

- **False Positives (FP):** A person does not have diabates but the model predicted person has diabetes. $\to$ $42$ FPs.

- **False Negatives (FN):** A person has diabetes but the model predicted person does not have diabetes. $\to$ $73$ FNs.
